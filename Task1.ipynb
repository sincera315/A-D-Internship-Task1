# Visualization

### Histograms

```python
sw=df.SepalWidthCm
sl=df.SepalLengthCm
pl=df.PetalLengthCm
pw=df.PetalWidthCm

plt.title("A histogram representing different histograms of numerical datas.")
plt.hist([sw,sl,pw,pl],orientation= "vertical", rwidth=5.5,color=['m','c','r','k'], label= ['SepalWidth','SepalLength','PetalLength','PetalWidth'])
plt.legend()
# these add x, y labels
plt.xlabel("Numerical Columns")
plt.ylabel("Ranges of their data in cm")
# for applying grid to visualization
plt.grid()
plt.show()
```

```python
# Providing the figure                                                                                                                                                                  size 
plt.figure(figsize=(10,12))
# -----------------------------------------------------
# Subplot for Sepal Width (top left)
plt.subplot(2, 2, 1)
plt.title("A histogram representing sample width.")
plt.hist(sw, histtype="bar", color='m')
plt.xlabel("Sample Width")
plt.ylabel("Ranges in cm")
plt.grid()

# -------------------------------------------------
# Subplot for Sepal Length (bottom left)
plt.subplot(2, 2, 3)
plt.title("A histogram representing sample length.")
plt.hist(sl, histtype="bar", color='c')
plt.xlabel("Sample Length")
plt.ylabel("Ranges in cm")
plt.grid()

# -------------------------------------------------
# Subplot for Petal Length (top right)
plt.subplot(2, 2, 2)
plt.title("A histogram representing petal length.")
plt.hist(pl, histtype="bar", color='r')
plt.xlabel("Petal Length")
plt.ylabel("Ranges in cm")
plt.grid()
# -------------------------------------------------

# Subplot for Petal Width (bottom right)
plt.subplot(2, 2, 4)
plt.title("A histogram representing petal width.")
plt.hist(pw, histtype="bar", color='k')
plt.xlabel("Petal Width")
plt.ylabel("Ranges in cm")
plt.grid()
# -------------------------------------------------

# Adjust spacing between subplots (optional)

plt.show()
```

### Scatter PLots

```python
# Providing the figure                                                                                                                                                                  size 
plt.figure(figsize=(10,12))
# Create scatter plots with labels and colors
plt.subplot(2, 2, 1)
plt.scatter(sw, sl, color='red', label='Sepal Width vs Sepal Length')

# Adding labels and title
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Sepal Length (cm)')
plt.title('Scatter Plot of Sepal Width vs Sepal Length')
# -------------------------------------------------
plt.subplot(2, 2, 2)
# Create scatter plots with labels and colors
plt.scatter(pw, pl, color='cyan')

# Adding labels and title
plt.xlabel('Petal Width (cm)')
plt.ylabel('Petal Length (cm)')
plt.grid()
plt.title('Scatter Plot of Petal width vs Petal Length')

# -------------------------------------------------
plt.subplot(2, 2, 3)
# Create scatter plots with labels and colors
plt.scatter(sw, pw, color='m', label='Petal Width vs Sepal Width')

# Adding labels and title
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Scatter Plot of Petal Width vs Sepal Width')

# -------------------------------------------------
plt.subplot(2, 2, 4)
# Create scatter plots with labels and colors
plt.scatter(sl, pl, color='y', label='Sepal length vs Petal Length')

# Adding labels and title
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.title('Scatter Plot of Petal Length vs Sepal Length')

# Show plot
plt.show()
```
